```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/Algerian_forest_fires_dataset.csv")
df
```

```python
df.drop([122,123,124],inplace=True)
df.reset_index(inplace=True)
df.drop('index',axis=1,inplace=True)
df
```

```python
df.loc[:'244','region']='region'
df.loc[:'244','region']='sidi-belie'
df
```

```python
df.columns=[i.strip() for i in df.columns ]
df.columns
df
```

```python
df.head()
```

```python
df.drop('Classes',axis=1,inplace=True)
```

```python
df
```

```python
# df.drop([122], axis=0, inplace=True)
df
```

```python
df['date']=pd.to_datetime(df[['day','month','year']])
df
```

```python
df.corr(numeric_only=True)
```

```python
df['Temperature']=df['Temperature'].astype(int)
df['RH']=df['RH'].astype(int)
df['Ws']=df['Ws'].astype(int)
df['Rain']=df['Rain'].astype(float)
df['FFMC']=df['FFMC'].astype(float)
df['DMC']=df['DMC'].astype(float)
df['ISI']=df['ISI'].astype(float)
df['BUI']=df['BUI'].astype(float)
df.dtypes
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
df['DC']=lb.fit_transform(df['DC'])
df['FWI']=lb.fit_transform(df['FWI'])
df['region']=lb.fit_transform(df['region'])
df.info()
```

```python
df
```

```python
x=pd.DataFrame(df,columns=['RH','Ws','Rain','FFMC','DMC','DC','ISI','BUI','FWI','region'])
y=pd.DataFrame(df,columns=['Temperature'])
y
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

```python
sns.displot(reg_pred-y_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(r2_score(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```
