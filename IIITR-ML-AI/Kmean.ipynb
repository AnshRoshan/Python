```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=100)
```

```python
x,y
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(x)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import  KMeans
```

# Elbow Method

```python
wcss=[]
for k in range(1,11):
  kmeans=KMeans(n_clusters=k,init='k-means++')
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
import scipy.cluster.hierarchy  as sc
plt.figure(figsize=(15,5))
plt.title("dendrogram")
sc.dendrogram(sc.linkage(x,method='ward'),distance_sort=True,)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("No. of cluster")
plt.ylabel("Wcss")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster= AgglomerativeClustering(n_clusters=11,affinity='euclidean',linkage='ward')
cluster.fit(x)
cluster.labels_
```
